## Survival Data
### Survival by Sex and Class

```python
survive_total = df.pivot_table(values=['survive', 'decease'], index=['sex','pclass'], aggfunc=[sum, np.mean],  margins_name='Total', margins=True)
survive_total
```

```python
survive_total.to_pickle('survival_by_class_sex')
```

### Sex Survival

```python
sex_survive = pd.crosstab(df['sex'] ,df['survived'])#,  margins=True)
sex_survive
```

```python
sex_survive.to_pickle('sex_survived_titanic')
```

```python
chisq_value_sex, pvalue_sex, dof_sex, expected_sex = chi2_contingency(sex_survive)
```

```python
#print 'expected', expected_sex

print 'chi score', chisq_value_sex
print 'degrees of freedom', dof_sex
print 'pvalue', pvalue_sex
```

### Class Survival

```python
class_survive = pd.crosstab(df['pclass'] ,df['survived'])#,  margins=True)
class_survive
```

```python
chisq_value, pvalue, dof, expected = chi2_contingency(class_survive)
```

```python
#print 'expected', expected

print 'chi score', chisq_value
print 'degrees of freedom', dof
print 'pvalue', pvalue
```
